```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
func = lambda x: 3*x**2 + 2*x +1
x = np.linspace(-2,2)
y = func(x)
```

```python
plt.plot(x,y)
```

```python
x = np.linspace(-np.pi,np.pi)
plt.plot(x,np.sin(x),label='sin')
plt.plot(x,np.cos(x),label='cos')
plt.legend()
```

```python
x = np.linspace(-1.3,1.3)
plt.plot(x,np.tan(x),label='tan')
```
